### Linear Layer:

$$y = x W^{T} + b$$

```python
torch.manual_seed(42)

linear = torch.nn.Linear(in_features=3, out_features=10, device=cuda, dtype=torch.float64)

# Since the matrix in the Linear layer is transposed, the in_features actually represent the number of columns of the input tensor.

# W has shape (out_features, in_features)

x = torch.tensor([
    [1, 2, 3],
    [3, 4, 55],
], device=cuda, dtype=torch.float64)

output = linear(x)
```

```python
print(f"Input shape: {x.shape}\n")
print(f"Output:\n\n{output}\n\nOutput shape: {output.shape}")
```

### Reshaping, Stacking, Squeezing & Unsqueezing:

```python
x = torch.arange(0, 8, dtype=torch.float32)
print(x)
print(x.shape)
```

```python
# copy the data

x_reshaped = x.reshape(2, 4)
print(x_reshaped)
print(x_reshaped.shape)
```

```python
# share the same memory, need to be continuous

z = x.view(2, 4)
print(z)
print(z.shape)
```

```python
x_stacked = torch.stack([x, x, x], dim = 0)
x_stacked
```

```python
x_stacked = torch.stack([x, x, x], dim = 1)
x_stacked
```

```python
x = x.reshape(1, 8)
x_squeezed = torch.squeeze(x)

x_unsqueezed = x_squeezed.unsqueeze(dim = 0)

print(f"{x} : {x.shape}")

print(f"{x_squeezed}   : {x_squeezed.shape}")

print(f"{x_unsqueezed} : {x_unsqueezed.shape}")
```

```python
# It is a view of the original -> if you change, the original also gets changed

x_original = torch.rand(size=(224,224,3))

x_permuted = x_original.permute(2, 0, 1)

print(x_original.shape)

print(x_permuted.shape)
```
